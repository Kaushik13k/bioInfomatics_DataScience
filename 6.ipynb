```python
import pandas as pd
diabetes=pd.read_csv('pima-indians-diabetes.csv')
print(list(diabetes))
X = diabetes.iloc[:, :8]
y = diabetes.iloc[:, 8]
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1, train_size=0.9, random_state=42)
print(X_train,X_test)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5)
#X=X.to_numpy()
#y=y.to_numpy()
i=1
for train_index, test_index in kf.split(X):
    print("TRAIN",i, ":", train_index, "TEST", i, ":", test_index)
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    i+=1
```

```python
from sklearn.model_selection import StratifiedKFold
kf = StratifiedKFold(n_splits=5)
#X=X.to_numpy()
#y=y.to_numpy()
i=1
for train_index, test_index in kf.split(X,y):
    print("TRAIN",i, ":", train_index, "TEST", i, ":", test_index)
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    i+=1
```

```python
#Random Forest model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=5, random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics
print("RF model accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import svm
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print("SVM model accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
mlp_model = MLPClassifier(solver='lbfgs', max_iter=10000,alpha=0.00001,hidden_layer_sizes=(5, 2), random_state=1)
mlp_model.fit(X_train, y_train)
y_pred = mlp_model.predict(X_test)
print("MLP model accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import tree
tree_model = tree.DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
y_pred = tree_model.predict(X_test)
print("Decision tree model accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
### Build model while setting probability to True.
from sklearn import svm
svm_model = svm.SVC(probability=True)
svm_model.fit(X_train, y_train)
### Build model while setting probability to True.
y_pred = svm_model.predict(X_test)
print("SVM model accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
y_pred = svm_model.predict(X_test)
y_pred_prob = svm_model.predict_proba(X_test)
print(y_pred)
print(y_pred_prob)
```

```python
print("F1 Score:",metrics.f1_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred, labels=[0, 1]))
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
print("tn: ",tn, "\nfp: ", fp, "\nfn: ", fn, "\ntp: ", tp)
#print(y_test.value_counts())
```

```python
import matplotlib.pyplot as plt
metrics.plot_roc_curve(svm_model, X_test, y_test)
plt.show()
```
